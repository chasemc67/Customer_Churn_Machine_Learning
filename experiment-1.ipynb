```python
from __future__ import division
import pandas as pd
import numpy as np

churn_df = pd.read_csv('Data/data.csv')
col_names = churn_df.columns.tolist()

print("Column names:")
print(col_names)

to_show = col_names[:6] + col_names[-6:]

print("\nSample data:")
churn_df[:].head(6)
```

```python
# Isolate target data
churn_result = churn_df['renewed']
y = churn_result.astype(int)
print(y)

# Drop the result from training data 
to_drop = ['renewed']
churn_df = churn_df.drop(to_drop,axis=1)

# Convert other T/F data to 1/0
yes_no_cols = ["annual_plan"]
churn_df[yes_no_cols] = churn_df[yes_no_cols].astype(int)

# Extract Features 
features = churn_df.columns

```

```python
country2int = dict(enumerate(set(churn_df['country'])))
industry2int = dict(enumerate(set(churn_df['industry'])))
category2int = dict(enumerate(set(churn_df['industry_category'])))
plancode2int = dict(enumerate(set(churn_df['plan_code'])))

int2country = {v: k for k, v in country2int.items()}
int2industry = {v: k for k, v in industry2int.items()}
int2category = {v: k for k, v in category2int.items()}
int2plancode = {v: k for k, v in plancode2int.items()}

churn_df['country'] = [int2country[i] for i in churn_df['country']]
churn_df['industry'] = [int2industry[i] for i in churn_df['industry']]
churn_df['industry_category'] = [int2category[i] for i in churn_df['industry_category']]
churn_df['plan_code'] = [int2plancode[i] for i in churn_df['plan_code']]

churn_df[:].head(6)
```

```python
#Convert data to np array 
X = churn_df.as_matrix().astype(np.float)
print(X)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X = scaler.fit_transform(X)

print("Feature space holds %d observations and %d features" % X.shape)
print("Unique target labels:", np.unique(y))
```
