```python
from __future__ import division
import pandas as pd
import numpy as np

churn_df = pd.read_csv('../Data/train_set.csv')
col_names = churn_df.columns.tolist()

print(type(churn_df))
to_show = col_names[:6] + col_names[-6:]

print("\nSample data:")
churn_df[:].head(6)
```

```python
# Isolate target data
churn_result = churn_df['renewed']
y = churn_result.astype(int)


# Drop the result from training data 
to_drop = ['renewed']
churn_df = churn_df.drop(to_drop,axis=1)

to_drop = ['cancellation_request']
churn_df = churn_df.drop(to_drop,axis=1)

to_drop = ['id']
churn_df = churn_df.drop(to_drop,axis=1)



# Extract Features 
features = churn_df.columns
# Identify True/False features 

tf_features = [col for col in features.tolist() if type(churn_df[col][0]) == np.bool_]

# Convert other T/F data to 1/0
churn_df[tf_features] = churn_df[tf_features].astype(int)



churn_df[:].head(6)
```

```python
country2int = dict(enumerate(set(churn_df['country'])))
industry2int = dict(enumerate(set(churn_df['industry'])))
category2int = dict(enumerate(set(churn_df['industry_category'])))
plancode2int = dict(enumerate(set(churn_df['plan_code'])))
cgroup2int = dict(enumerate(set(churn_df['campaign_group'])))
cstr2int = dict(enumerate(set(churn_df['campaign_string'])))

int2country = {v: k for k, v in country2int.items()}
int2industry = {v: k for k, v in industry2int.items()}
int2category = {v: k for k, v in category2int.items()}
int2plancode = {v: k for k, v in plancode2int.items()}
int2cgroup = {v: k for k, v in cgroup2int.items()}
int2cstr = {v: k for k, v in cstr2int.items()}

churn_df['country'] = [int2country[i] for i in churn_df['country']]
churn_df['industry'] = [int2industry[i] for i in churn_df['industry']]
churn_df['industry_category'] = [int2category[i] for i in churn_df['industry_category']]
churn_df['plan_code'] = [int2plancode[i] for i in churn_df['plan_code']]
churn_df['campaign_group'] = [int2cgroup[i] for i in churn_df['campaign_group']]
churn_df['campaign_string'] = [int2cstr[i] for i in churn_df['campaign_string']]

churn_df['account_value'] = churn_df['account_value'].replace(np.nan, 0)
churn_df = churn_df.replace(np.nan, 100000000)
churn_df[:].head(6)



```

```python
#Convert data to np array 
X = churn_df.as_matrix().astype(np.float)
print(X)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X = scaler.fit_transform(X)

print("Feature space holds %d observations and %d features" % X.shape)
print("Unique target labels:", np.unique(y))
```

```python
from sklearn.cross_validation import KFold

def run_cv(X,y,clf_class,**kwargs):
    # Construct a kfolds object
    kf = KFold(len(y),n_folds=5,shuffle=True)
    y_pred = y.copy()

    # Iterate through folds
    for train_index, test_index in kf:
        X_train, X_test = X[train_index], X[test_index]
        y_train = y[train_index]
        # Initialize a classifier with key word arguments
        clf = clf_class(**kwargs)
        clf.fit(X_train,y_train)
        y_pred[test_index] = clf.predict(X_test)
    return y_pred
```

```python
from sklearn.svm import SVC
from sklearn.svm import NuSVC

def accuracy(y_true,y_pred):
    # NumPy interprets True and False as 1. and 0.
    return np.mean(y_true == y_pred)

print("Support vector machines:")
print("%.3f" % accuracy(y, run_cv(X,y,SVC)))

print("Nu-SVMs:")
print("%.3f" % accuracy(y, run_cv(X,y,NuSVC)))
```
